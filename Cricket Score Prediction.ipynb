```python
#venue changed to int
df['venue'] = df.venue.astype(int)
df.info()
```

```python
#converting date dtype from object to datetime format
df['date']=df['date'].apply(lambda x:datetime.strptime(x,'%Y-%m-%d'))
```

```python
#Removing Unnecessary Columns
df.drop(['mid','batsman','bowler','striker','non-striker'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#checking for the overall teams present in bat_team
df['bat_team'].unique()
```

```python
#checking for the overall teams present in bowl_team
df['bowl_team'].unique()
```

```python
#Grabbing the standard exsisted teams only
standard_teams=['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals','Mumbai Indians', 'Kings XI Punjab',
                'Royal Challengers Bangalore', 'Delhi Daredevils', 'Sunrisers Hyderabad']
```

```python
#Forming the dataframe with standard_teams list
df=df[df['bat_team'].isin(standard_teams)]
df=df[df['bowl_team'].isin(standard_teams)]
```

```python
df
```

```python
#checking for unique venues
df['venue'].unique()
```

```python
#number of unique venues 
df['venue'].nunique()
```

```python
#venue is a ordinal categorical variable and ranks are alocatted in the previous step!
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['venue']=le.fit_transform(df['venue'].values)
df
```

```python
#minimum of 5 overs score is necessary in order to predict the score!
df=df[df['overs']>=5.0]
df
```

```python
#reset the index
df=df.reset_index(drop=True)
df
```

```python
#bat_team and bowl_team are nominal categorical variables, so one_hot_encoding is performed
one_hot_encoded_df=pd.get_dummies(df[['bat_team','bowl_team']])
one_hot_encoded_df
```

```python
#Encoding is performed in the above step! SO,now there is no use with bat_team and bowl_team columns. Drop them!
df.drop(df[['bat_team','bowl_team']],axis=1,inplace=True)
```

```python
df
```

```python
#merge the df and one_hot_encoded_df
df=pd.merge(left=df,right=one_hot_encoded_df,how='outer',left_index=True,right_index=True)
df
```

```python
#columns of df
df.columns
```

```python
#changing the position of the columns, in order to shift the total column to last.(It might be easy to keep total column as dependent variable
df=pd.DataFrame(data=df,columns=['date', 'venue', 'runs', 'wickets', 'overs', 'runs_last_5',
       'wickets_last_5', 'bat_team_Chennai Super Kings',
       'bat_team_Delhi Daredevils', 'bat_team_Kings XI Punjab',
       'bat_team_Kolkata Knight Riders', 'bat_team_Mumbai Indians',
       'bat_team_Rajasthan Royals', 'bat_team_Royal Challengers Bangalore',
       'bat_team_Sunrisers Hyderabad', 'bowl_team_Chennai Super Kings',
       'bowl_team_Delhi Daredevils', 'bowl_team_Kings XI Punjab',
       'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians',
       'bowl_team_Rajasthan Royals', 'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad','total'])
```

```python
df
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
```

```python
#split dataset into X_train,X_test,y_train,y_test
X_train=df.iloc[:,:-1][df['date'].dt.year<=2015]
X_test=df.iloc[:,:-1][df['date'].dt.year>=2016]
```

```python
y_train=df[df['date'].dt.year<=2015].iloc[:,-1]
y_test=df[df['date'].dt.year>=2016].iloc[:,-1]
```

```python
#the dataset is splitted into test and train sets with the help of date column, and now there is no use with it.So,drop the date column!
X_train.drop(labels='date', axis=True, inplace=True)
X_test.drop(labels='date', axis=True, inplace=True)
```

```python
#LINEAR_Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
#pass the X_test to predict the y values
y_pred=lr.predict(X_test)
y_pred
```

```python
#Find the accuracy between X_test and y_test
lr.score(X_test,y_test)
```

```python
#Find MSE,RMSE,MAE
from sklearn import metrics
import math
mean_square_error=metrics.mean_squared_error(y_pred,y_test)
root_mean_square_error=np.sqrt(metrics.mean_squared_error(y_pred,y_test))
mean_absolute_error=metrics.mean_absolute_error(y_pred,y_test)
```

```python
print("MSE:",mean_square_error,"\n" "RMSE:",root_mean_square_error,"\n" ,"MAE:",mean_absolute_error)
```

```python
#R-SQUARED ERROR
SSres=sum((y_pred-y_test))**2
SStotal=sum((y_test.mean()-y_test))**2
r2_error=(SSres/SStotal)
r2_error
```
